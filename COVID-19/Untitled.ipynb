```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import os, glob
```

```python
ext = 'csv'
os.chdir('../COVID-19/Covid_canada_data')
files = glob.glob('*.{}'.format(ext))
```

```python
files
```

```python
files_dict={}
name_list = []
for file in files:
    files_dict[os.path.basename(file).split('.')[0]] = pd.read_csv(file)
    name_list.append(os.path.basename(file).split('.')[0])
```

```python
for file in name_list:
    print(file)
    display(files_dict[file].head())
    display(files_dict[file].info())
```
